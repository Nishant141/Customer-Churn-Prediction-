```python
# Google Colab PySpark End-to-End E-commerce Analytics Project
# Direct Kaggle data access without manual download

# ==================== SETUP & INSTALLATION ====================

# Install required packages
!pip install pyspark
!pip install kaggle
!pip install plotly
!pip install seaborn

# Install Java (required for PySpark)
!apt-get update -qq
!apt-get install openjdk-8-jdk -qq
import os
os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-8-openjdk-amd64"

# ==================== KAGGLE SETUP ====================

# Upload your kaggle.json file using the file upload widget
from google.colab import files
print("/content/kaggle.json")
uploaded = files.upload()

# Set up Kaggle credentials
!mkdir -p ~/.kaggle
!mv kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json

# Verify Kaggle setup
!kaggle --version

# ==================== DATA ACQUISITION ====================

# Download Brazilian E-Commerce dataset directly
!kaggle datasets download -d olistbr/brazilian-ecommerce
!unzip -o brazilian-ecommerce.zip

# List downloaded files
!ls -la *.csv

# ==================== PYSPARK INITIALIZATION ====================

import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.ml.feature import VectorAssembler, StandardScaler, StringIndexer
from pyspark.ml.clustering import KMeans
from pyspark.ml.recommendation import ALS
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import ClusteringEvaluator

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Create Spark Session
spark = SparkSession.builder \
    .appName("E-commerce Analytics Pipeline") \
    .config("spark.sql.adaptive.enabled", "true") \
    .config("spark.sql.adaptive.coalescePartitions.enabled", "true") \
    .getOrCreate()

print(f"Spark version: {spark.version}")
print(f"Spark UI available at: {spark.sparkContext.uiWebUrl}")

# ==================== DATA LOADING & EXPLORATION ====================

# Load all datasets
orders_df = spark.read.option("header", "true").option("inferSchema", "true").csv("olist_orders_dataset.csv")
order_items_df = spark.read.option("header", "true").option("inferSchema", "true").csv("olist_order_items_dataset.csv")
customers_df = spark.read.option("header", "true").option("inferSchema", "true").csv("olist_customers_dataset.csv")
products_df = spark.read.option("header", "true").option("inferSchema", "true").csv("olist_products_dataset.csv")
sellers_df = spark.read.option("header", "true").option("inferSchema", "true").csv("olist_sellers_dataset.csv")
reviews_df = spark.read.option("header", "true").option("inferSchema", "true").csv("olist_order_reviews_dataset.csv")

print("=== DATASET OVERVIEW ===")
datasets = {
    "Orders": orders_df,
    "Order Items": order_items_df,
    "Customers": customers_df,
    "Products": products_df,
    "Sellers": sellers_df,
    "Reviews": reviews_df
}

for name, df in datasets.items():
    print(f"{name}: {df.count()} rows, {len(df.columns)} columns")

# Show sample data
print("\n=== ORDERS SAMPLE ===")
orders_df.show(5, truncate=False)

print("\n=== ORDER ITEMS SAMPLE ===")
order_items_df.show(5, truncate=False)

# ==================== DATA CLEANING & PREPROCESSING ====================

# Clean orders data
orders_cleaned = orders_df \
    .filter(col("order_status") == "delivered") \
    .filter(col("order_delivered_customer_date").isNotNull()) \
    .withColumn("order_purchase_timestamp", to_timestamp(col("order_purchase_timestamp"))) \
    .withColumn("order_delivered_customer_date", to_timestamp(col("order_delivered_customer_date"))) \
    .withColumn("delivery_days",
                datediff(col("order_delivered_customer_date"), col("order_purchase_timestamp"))) \
    .filter(col("delivery_days") >= 0)

print(f"Orders after cleaning: {orders_cleaned.count()} rows")

# Clean order items and add calculations
order_items_cleaned = order_items_df \
    .filter(col("price") > 0) \
    .filter(col("freight_value") >= 0) \
    .withColumn("total_item_value", col("price") + col("freight_value"))

print(f"Order items after cleaning: {order_items_cleaned.count()} rows")

# ==================== COMPREHENSIVE DATA ANALYSIS ====================

# 1. ORDER ANALYSIS
print("\n=== ORDER ANALYSIS ===")

# Join orders with order items for comprehensive analysis
order_analysis = orders_cleaned.alias("o") \
    .join(order_items_cleaned.alias("oi"), col("o.order_id") == col("oi.order_id")) \
    .select(
        col("o.order_id"),
        col("o.customer_id"),
        col("o.order_purchase_timestamp"),
        col("o.delivery_days"),
        col("oi.product_id"),
        col("oi.seller_id"),
        col("oi.price"),
        col("oi.freight_value"),
        col("oi.total_item_value")
    )

# Monthly sales trends
monthly_sales = order_analysis \
    .withColumn("year_month", date_format(col("order_purchase_timestamp"), "yyyy-MM")) \
    .groupBy("year_month") \
    .agg(
        sum("total_item_value").alias("total_revenue"),
        count("order_id").alias("total_orders"),
        countDistinct("customer_id").alias("unique_customers")
    ) \
    .orderBy("year_month")

monthly_sales.show(20)

# 2. CUSTOMER SEGMENTATION ANALYSIS
print("\n=== CUSTOMER SEGMENTATION ===")

# Calculate customer metrics
customer_metrics = order_analysis \
    .groupBy("customer_id") \
    .agg(
        sum("total_item_value").alias("total_spent"),
        count("order_id").alias("order_frequency"),
        countDistinct("product_id").alias("product_variety"),
        avg("total_item_value").alias("avg_order_value"),
        max("order_purchase_timestamp").alias("last_purchase_date"),
        min("order_purchase_timestamp").alias("first_purchase_date"),
        avg("delivery_days").alias("avg_delivery_days")
    ) \
    .withColumn("customer_lifetime_days",
                datediff(col("last_purchase_date"), col("first_purchase_date"))) \
    .filter(col("total_spent") > 0)

print(f"Customer metrics calculated for {customer_metrics.count()} customers")
customer_metrics.describe().show()

# Prepare data for clustering
feature_columns = ["total_spent", "order_frequency", "product_variety", "avg_order_value"]

# Remove outliers using IQR method for each feature
def remove_outliers_iqr(df, columns):
    for column in columns:
        quantiles = df.approxQuantile(column, [0.25, 0.75], 0.05)
        q1, q3 = quantiles[0], quantiles[1]
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        df = df.filter((col(column) >= lower_bound) & (col(column) <= upper_bound))
    return df

customer_metrics_clean = remove_outliers_iqr(customer_metrics, feature_columns)
print(f"After outlier removal: {customer_metrics_clean.count()} customers")

# K-Means Clustering for Customer Segmentation
assembler = VectorAssembler(
    inputCols=feature_columns,
    outputCol="features"
)

scaler = StandardScaler(
    inputCol="features",
    outputCol="scaled_features"
)

# Find optimal K using elbow method
def find_optimal_k(df, max_k=10):
    costs = []
    for k in range(2, max_k + 1):
        kmeans = KMeans(k=k, featuresCol="scaled_features", predictionCol="cluster")
        pipeline = Pipeline(stages=[assembler, scaler, kmeans])
        model = pipeline.fit(df)
        predictions = model.transform(df)

        evaluator = ClusteringEvaluator(predictionCol="cluster", featuresCol="scaled_features")
        cost = evaluator.evaluate(predictions)
        costs.append((k, cost))
        print(f"K={k}, Silhouette Score={cost:.4f}")

    return costs

print("Finding optimal number of clusters...")
k_costs = find_optimal_k(customer_metrics_clean, max_k=8)

# Use K=4 for customer segmentation (typically good for business interpretation)
kmeans = KMeans(k=4, featuresCol="scaled_features", predictionCol="customer_segment", seed=42)
clustering_pipeline = Pipeline(stages=[assembler, scaler, kmeans])
clustering_model = clustering_pipeline.fit(customer_metrics_clean)

# Apply clustering
customer_segments = clustering_model.transform(customer_metrics_clean)

print("\n=== CUSTOMER SEGMENTS ===")
segment_summary = customer_segments.groupBy("customer_segment") \
    .agg(
        count("*").alias("segment_size"),
        avg("total_spent").alias("avg_total_spent"),
        avg("order_frequency").alias("avg_order_frequency"),
        avg("product_variety").alias("avg_product_variety"),
        avg("avg_order_value").alias("avg_order_value")
    ) \
    .orderBy("customer_segment")

segment_summary.show()

# 3. PRODUCT ANALYSIS
print("\n=== PRODUCT ANALYSIS ===")

# Join products with order items for product analysis
product_analysis = order_items_cleaned.alias("oi") \
    .join(products_df.alias("p"), col("oi.product_id") == col("p.product_id")) \
    .select(
        col("oi.product_id"),
        col("p.product_category_name"),
        col("oi.price"),
        col("oi.freight_value"),
        col("oi.total_item_value")
    )

# Top product categories by revenue
top_categories = product_analysis \
    .groupBy("product_category_name") \
    .agg(
        sum("total_item_value").alias("total_revenue"),
        count("*").alias("total_sales"),
        avg("price").alias("avg_price")
    ) \
    .orderBy(col("total_revenue").desc())

print("Top 15 Product Categories by Revenue:")
top_categories.show(15)

# Product performance metrics
product_performance = product_analysis \
    .groupBy("product_id", "product_category_name") \
    .agg(
        sum("total_item_value").alias("total_revenue"),
        count("*").alias("sales_count"),
        avg("price").alias("avg_price")
    ) \
    .orderBy(col("total_revenue").desc())

print("Top 10 Products by Revenue:")
product_performance.show(10)

# 4. RECOMMENDATION SYSTEM
print("\n=== BUILDING RECOMMENDATION SYSTEM ===")

# Prepare data for collaborative filtering
# Create implicit ratings based on purchase behavior
ratings_data = order_analysis \
    .groupBy("customer_id", "product_id") \
    .agg(
        count("*").alias("purchase_count"),
        sum("total_item_value").alias("total_spent")
    ) \
    .withColumn("rating",
                when(col("purchase_count") >= 3, 5.0)
                .when(col("purchase_count") == 2, 4.0)
                .otherwise(3.0 + (col("total_spent") / 100).cast("double"))) \
    .filter(col("rating") <= 5.0)

print(f"Rating data prepared: {ratings_data.count()} user-item interactions")

# Convert string IDs to numeric for ALS
customer_indexer = StringIndexer(inputCol="customer_id", outputCol="customer_index")
product_indexer = StringIndexer(inputCol="product_id", outputCol="product_index")

indexed_ratings = customer_indexer.fit(ratings_data).transform(ratings_data)
indexed_ratings = product_indexer.fit(indexed_ratings).transform(indexed_ratings)

# Train ALS model
als = ALS(
    userCol="customer_index",
    itemCol="product_index",
    ratingCol="rating",
    maxIter=10,
    regParam=0.1,
    rank=50,
    coldStartStrategy="drop",
    nonnegative=True,
    seed=42
)

# Split data for training and validation
train_data, test_data = indexed_ratings.randomSplit([0.8, 0.2], seed=42)

als_model = als.fit(train_data)

print("ALS Recommendation model trained successfully!")

# Generate recommendations for all users
user_recommendations = als_model.recommendForAllUsers(5)
print("Sample user recommendations:")
user_recommendations.show(5, truncate=False)

# 5. BUSINESS INSIGHTS & KPIs
print("\n=== KEY BUSINESS INSIGHTS ===")

# Calculate key business metrics
total_revenue = order_analysis.agg(sum("total_item_value")).collect()[0][0]
total_orders = order_analysis.count()
unique_customers = order_analysis.select("customer_id").distinct().count()
unique_products = order_analysis.select("product_id").distinct().count()
avg_order_value = total_revenue / total_orders

print(f"ðŸ“Š BUSINESS KPIs:")
print(f"   ðŸ’° Total Revenue: ${total_revenue:,.2f}")
print(f"   ðŸ“¦ Total Orders: {total_orders:,}")
print(f"   ðŸ‘¥ Unique Customers: {unique_customers:,}")
print(f"   ðŸ›ï¸  Unique Products: {unique_products:,}")
print(f"   ðŸ’µ Average Order Value: ${avg_order_value:.2f}")

# Customer acquisition by month
customer_acquisition = orders_cleaned \
    .groupBy("customer_id") \
    .agg(min("order_purchase_timestamp").alias("first_purchase")) \
    .withColumn("acquisition_month", date_format(col("first_purchase"), "yyyy-MM")) \
    .groupBy("acquisition_month") \
    .count() \
    .orderBy("acquisition_month")

print("\nðŸ“ˆ Monthly Customer Acquisition:")
customer_acquisition.show()

# ==================== VISUALIZATION (Convert to Pandas for plotting) ====================

print("\n=== CREATING VISUALIZATIONS ===")

# Convert key datasets to Pandas for visualization
monthly_sales_pd = monthly_sales.toPandas()
segment_summary_pd = segment_summary.toPandas()
top_categories_pd = top_categories.limit(10).toPandas()

# Set up plotting style
plt.style.use('seaborn-v0_8')
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Monthly Sales Trend
axes[0, 0].plot(monthly_sales_pd['year_month'], monthly_sales_pd['total_revenue'],
                marker='o', linewidth=2, markersize=6)
axes[0, 0].set_title('Monthly Revenue Trend', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Month')
axes[0, 0].set_ylabel('Revenue ($)')
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 0].grid(True, alpha=0.3)

# 2. Customer Segments
axes[0, 1].bar(segment_summary_pd['customer_segment'], segment_summary_pd['segment_size'])
axes[0, 1].set_title('Customer Segments Distribution', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Segment')
axes[0, 1].set_ylabel('Number of Customers')
axes[0, 1].grid(True, alpha=0.3)

# 3. Top Product Categories
axes[1, 0].barh(top_categories_pd['product_category_name'][:8],
                top_categories_pd['total_revenue'][:8])
axes[1, 0].set_title('Top Product Categories by Revenue', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Revenue ($)')
axes[1, 0].grid(True, alpha=0.3)

# 4. Average Order Value by Segment
avg_order_by_segment = segment_summary_pd[['customer_segment', 'avg_order_value']]
axes[1, 1].bar(avg_order_by_segment['customer_segment'], avg_order_by_segment['avg_order_value'])
axes[1, 1].set_title('Average Order Value by Customer Segment', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Customer Segment')
axes[1, 1].set_ylabel('Average Order Value ($)')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ==================== ADVANCED ANALYTICS ====================

print("\n=== ADVANCED ANALYTICS ===")

# Cohort Analysis
def create_cohort_analysis():
    # Get first purchase date for each customer
    customer_first_purchase = orders_cleaned \
        .groupBy("customer_id") \
        .agg(min("order_purchase_timestamp").alias("first_purchase_date"))

    # Join with all orders to create cohort data
    cohort_data = orders_cleaned.alias("o") \
        .join(customer_first_purchase.alias("c"), col("o.customer_id") == col("c.customer_id")) \
        .withColumn("cohort_month", date_format(col("first_purchase_date"), "yyyy-MM")) \
        .withColumn("purchase_month", date_format(col("order_purchase_timestamp"), "yyyy-MM")) \
        .withColumn("months_since_first_purchase",
                   months_between(col("order_purchase_timestamp"), col("first_purchase_date")))

    return cohort_data

cohort_data = create_cohort_analysis()
print("Cohort analysis data created")

# RFM Analysis (Recency, Frequency, Monetary)
current_date = orders_cleaned.agg(max("order_purchase_timestamp")).collect()[0][0]

rfm_analysis = customer_metrics \
    .withColumn("recency_days", datediff(lit(current_date), col("last_purchase_date"))) \
    .select(
        col("customer_id"),
        col("recency_days").alias("recency"),
        col("order_frequency").alias("frequency"),
        col("total_spent").alias("monetary")
    )

# Calculate RFM scores (1-5 scale)
rfm_quantiles = rfm_analysis.approxQuantile("recency", [0.2, 0.4, 0.6, 0.8], 0.05)
frequency_quantiles = rfm_analysis.approxQuantile("frequency", [0.2, 0.4, 0.6, 0.8], 0.05)
monetary_quantiles = rfm_analysis.approxQuantile("monetary", [0.2, 0.4, 0.6, 0.8], 0.05)

rfm_scores = rfm_analysis \
    .withColumn("r_score",
        when(col("recency") <= rfm_quantiles[0], 5)
        .when(col("recency") <= rfm_quantiles[1], 4)
        .when(col("recency") <= rfm_quantiles[2], 3)
        .when(col("recency") <= rfm_quantiles[3], 2)
        .otherwise(1)) \
    .withColumn("f_score",
        when(col("frequency") <= frequency_quantiles[0], 1)
        .when(col("frequency") <= frequency_quantiles[1], 2)
        .when(col("frequency") <= frequency_quantiles[2], 3)
        .when(col("frequency") <= frequency_quantiles[3], 4)
        .otherwise(5)) \
    .withColumn("m_score",
        when(col("monetary") <= monetary_quantiles[0], 1)
        .when(col("monetary") <= monetary_quantiles[1], 2)
        .when(col("monetary") <= monetary_quantiles[2], 3)
        .when(col("monetary") <= monetary_quantiles[3], 4)
        .otherwise(5)) \
    .withColumn("rfm_score", concat(col("r_score"), col("f_score"), col("m_score")))

print("RFM Analysis completed")
rfm_scores.show(10)

# ==================== MODEL PERSISTENCE & DEPLOYMENT READY CODE ====================

print("\n=== MODEL PERSISTENCE ===")

# Save models and processed data
try:
    # Save clustering model
    clustering_model.write().overwrite().save("/content/customer_segmentation_model")

    # Save recommendation model
    als_model.write().overwrite().save("/content/recommendation_model")

    # Save processed datasets
    customer_segments.write().mode("overwrite").parquet("/content/customer_segments")
    product_performance.write().mode("overwrite").parquet("/content/product_performance")
    rfm_scores.write().mode("overwrite").parquet("/content/rfm_analysis")

    print("âœ… All models and datasets saved successfully!")

except Exception as e:
    print(f"âŒ Error saving models: {e}")

# ==================== FINAL BUSINESS RECOMMENDATIONS ====================

print("\n" + "="*60)
print("ðŸŽ¯ FINAL BUSINESS RECOMMENDATIONS")
print("="*60)

# Customer segment insights
segment_insights = segment_summary.collect()
for row in segment_insights:
    segment = row['customer_segment']
    size = row['segment_size']
    avg_spent = row['avg_total_spent']
    avg_frequency = row['avg_order_frequency']

    print(f"\nðŸ“Š SEGMENT {segment}:")
    print(f"   Size: {size} customers ({size/unique_customers*100:.1f}%)")
    print(f"   Avg Spent: ${avg_spent:.2f}")
    print(f"   Avg Orders: {avg_frequency:.1f}")

    if avg_spent > total_revenue/unique_customers * 1.5:
        print("   ðŸ’Ž HIGH VALUE - Focus on retention")
    elif avg_frequency < 2:
        print("   ðŸŽ¯ LOW ENGAGEMENT - Target with promotions")
    else:
        print("   âš¡ GROWTH POTENTIAL - Upsell opportunities")

print(f"\nðŸš€ NEXT STEPS:")
print(f"   1. Deploy recommendation system for personalized marketing")
print(f"   2. Create targeted campaigns for each customer segment")
print(f"   3. Focus on top {unique_products//10} products for inventory optimization")
print(f"   4. Implement retention strategies for high-value customers")
print(f"   5. Set up real-time dashboards for monitoring KPIs")

print("\nâœ¨ PySpark E-commerce Analytics Pipeline Complete! âœ¨")

# Clean up
spark.stop()
```
